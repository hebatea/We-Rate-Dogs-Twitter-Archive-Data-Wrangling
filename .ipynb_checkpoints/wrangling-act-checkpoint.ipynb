## Introduction
In this project we will wrangle and analyze the tweet archive of Twitter user **@dog_rates**, also known as **WeRateDogs**. **WeRateDogs** is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. WeRateDogs asks people to send photos of their dogs which are rated on a scale of one to ten, but are invariably given ratings in excess of the maximum, such as "13/10".

#### `twitter_archive`: Missing values in (expanded_urls) column.

##### Define

Loop over the missing values of expanded urls, and because the tweet_id is the last part of the tweet URL after "status/", make the expanded urls = https:\//twitter.com\/dog_rates\/status\/tweet_id

##### Code

```python
# Identify the variables
length = archive_clean.loc[archive_clean.expanded_urls.isnull(), 'expanded_urls'].shape[0]
mtweet_ids = []
urls = []

# 
for i in range(0, length):
    mtweet_id = archive_clean.loc[archive_clean.expanded_urls.isnull(), 'tweet_id'].values[i]
    mtweet_ids.append(mtweet_id)
    urls.append("https://twitter.com/dog_rates/status/{}".format(mtweet_id))

for i in range(0, length):    
     archive_clean.loc[(archive_clean.tweet_id == mtweet_ids[i]), 'expanded_urls'] = urls[i]
```

### Research Question 1 : What is the relationship between *retweet_count* and *favorite_count*

We can answer this question using linear regression, and scatter plot as the following:

```python
twitter_archive_cleaned['intercept'] = 1

#OLS stands for Ordinary Least Square-> first we provide y, and then the x value
lm = sm.OLS(twitter_archive_cleaned['favorite_count'], twitter_archive_cleaned[['intercept','retweet_count']])
results = lm.fit()
results.summary() 
```

```python
#plot the relationship between retweet_count and favorite_count
twitter_archive_cleaned.plot(kind='scatter', x='retweet_count', y='favorite_count', title='relationship between retweet_count and favorite_count');
```

**From the model summary and scatter plot we can conclude the following:**

- The predicated favorite_count = 2485.5456 + 2.2721 * retweet_count -> if the retweet_count = 0, we predicate that the favorite_count will be around 2486, and for every one more retweet, we predicate that the favorite_count would be increase by 2.2721.

- The p-value for retweet_count = 0 -> retweet_count is statistically significant for predicting the favorite_count.

- The R-squared (= 0.833) is closer to 1, which mean the better our model fit, and it is suggest that there is a positive strong realtionship between favorite_count and retweet_count (as you can see in the scatterplot). Also We can interpret the value of R-squard as that 83.3% of the variability in favorite_count is explained by the retweet_count.

### Research Question 2 : Is there a relationship between *soure* and *favorite_count*

```python
# To get the names of dummies column
twitter_archive_cleaned.source.value_counts()
```

```python
# We always drop one level from each category. This is called the baseline
# Here the baseline is Vine - Make a Scene 
twitter_archive_cleaned[['iphone', 'web_client', 'tweetdeck']] = pd.get_dummies(twitter_archive_cleaned['source'])[['Twitter for iPhone', 'Twitter Web Client', 'TweetDeck']]

# We already define the intercept before
lm = sm.OLS(twitter_archive_cleaned.favorite_count, twitter_archive_cleaned[['intercept', 'iphone', 'web_client', 'tweetdeck']])
results = lm.fit()
results.summary()
```

**Conculsion :** <br>
- Based on Multiple Linear Regression : If our tweets are from *vine* we predict its favorite_count to be 8,034, the tweets from *iPhone* 1,286 greater than *vine*,  a *web_client* 2,332 less than *vine*, a *tweetdeck* 2,558 than *vine*.

- Based on plot: we can notice that if we tweets from iPhone the means of favorite_count will be higher.

### Research Question 6:  What is the relationship between *favorite_count* and *rating*?

```python
rating_num_median = twitter_archive_cleaned.rating_numerator.median()
rating_num_median
```

```python
higher_than_median = twitter_archive_cleaned.query('rating_numerator > {}'.format(rating_num_median))
higher_than_median['favorite_count'].mean(), higher_than_median['retweet_count'].mean()
```

```python
lower_than_median = twitter_archive_cleaned.query('rating_numerator <= {}'.format(rating_num_median))
lower_than_median['favorite_count'].mean(), lower_than_median['retweet_count'].mean()
```

**Conclusion**
- When the rating was > (median = 11) the mean of favorite_count was = 15,936, and the retweet_count = 4,934
- When the rating was <= (median = 11) the mean of favorite_count was = 4,685, and the retweet_count = 1,625
- The Higher rating go along with higher favorite_count and retweet_count
