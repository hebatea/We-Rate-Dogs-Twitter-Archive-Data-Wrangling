```python
with open('./Data/tweet-json.txt') as tweet_json:
    df_list = []
    for line in tweet_json:
        json_line = json.loads(line)
        tweet_id = json_line['id']
        retweet_count = json_line['retweet_count']
        favorite_count = json_line['favorite_count']
        df_list.append({
            'tweet_id' : tweet_id,
            'retweet_count' : retweet_count,
            'favorite_count' : favorite_count,
        })
        
```

```python
df_json = pd.DataFrame(df_list, columns = ['tweet_id', 'retweet_count', 'favorite_count'])
```

##### Define

Filter out the tweets that are not reply to anther, and filter out the tweets that are not reply to anther, and 

##### Code

```python
#Define some variables
in_reply_to_status_id = archive_clean.in_reply_to_status_id
in_reply_to_user_id = archive_clean.in_reply_to_user_id
retweeted_status_id = archive_clean.retweeted_status_id
retweeted_status_user_id = archive_clean.retweeted_status_user_id
retweeted_status_timestamp = archive_clean.retweeted_status_timestamp

# Filter out the tweets that are not reply to anther
archive_clean = archive_clean[in_reply_to_status_id.isnull() 
                              & in_reply_to_user_id.isnull()]

# Filter out the tweets that are not retweet
archive_clean  = archive_clean[retweeted_status_id.isnull()
                                & retweeted_status_user_id.isnull()
                                & retweeted_status_timestamp.isnull()]
```

#### `twitter_archive`: Missing values in (expanded_urls) column.

##### Define

Loop over the missing values of expanded urls, and because the tweet_id is the last part of the tweet URL after "status/", make the expanded urls = https:\//twitter.com\/dog_rates\/status\/tweet_id

##### Code

```python
# Identify the variables
length = archive_clean.loc[archive_clean.expanded_urls.isnull(), 'expanded_urls'].shape[0]
mtweet_ids = []
urls = []

# 
for i in range(0, length):
    mtweet_id = archive_clean.loc[archive_clean.expanded_urls.isnull(), 'tweet_id'].values[i]
    mtweet_ids.append(mtweet_id)
    urls.append("https://twitter.com/dog_rates/status/{}".format(mtweet_id))

for i in range(0, length):    
     archive_clean.loc[(archive_clean.tweet_id == mtweet_ids[i]), 'expanded_urls'] = urls[i]
```
